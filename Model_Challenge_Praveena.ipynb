```python
#Import pandas and numpy libraries to read and manipulate dataset
import pandas as pd
import numpy as np
```

```python
#Read the file from desktop
df = pd.read_csv("C:/Users/praveena/Desktop/Bank_Project/dsm-bank-model-2870047/bankData/bank.csv")
df.head()
```

```python
#Dropping campaign related columns and recoginsing features and output columns
df1 = df.copy()
df1.drop(df1.iloc[:, 8:16], inplace = True, axis = 1)
df1.head()
```

```python
X = df.iloc[:, :-1].values               #Feature columns
y = df.iloc[:, -1].values                #Output variable
```

```python
#Importing sklearn libraries for data pre-processing and splitting data to training and testing sets
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df1.columns
```

```python
'''
   Recognising datatypes and naming columns to give processed data which is easy to model.
   First let's work on numerical columns and then categorical columns, from 'df' dataframe, 
   column numbers 0, 5 are numeric while others are categorical datatype.
'''
numeric_data = df1.iloc[:, [0, 5]].values
numeric_df = pd.DataFrame(numeric_data, dtype = object)
numeric_df.columns = ['age', 'balance']
numeric_df.head()
```

```python
X_categoric = df1.iloc[:, [1,2,3,4,6,7]].values
```

```python
#applying StandardScaler to 'numeric_df' as both columns have values with a varied range
age_std_scale = StandardScaler()
numeric_df['age'] = age_std_scale.fit_transform(numeric_df[['age']])
```

```python
balance_std_scale = StandardScaler()
numeric_df['balance'] = balance_std_scale.fit_transform(numeric_df[['balance']])
numeric_df.head()
```

```python
#To apply ML models we need to convert categorical variables to numeric variables which can be done through "One-hot encoding".
one = OneHotEncoder()
categoric_data = one.fit_transform(X_categoric).toarray()
categoric_df = pd.DataFrame(categoric_data)
categoric_df.columns = one.get_feature_names_out()
categoric_df
```

```python
#Now let's concatenate two dataframes
X_final = pd.concat([numeric_df, categoric_df], axis = 1)
```

```python
#Using RandomForestClassifier as a ML model for data modeling
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 100)
rfc.fit(X_final, y)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X_final,y, test_size= 0.2)
rfc = RandomForestClassifier(n_estimators = 100)
rfc.fit(X_train, y_train)
```

```python
y_pred=rfc.predict(X_test)
from sklearn.metrics import confusion_matrix
confusion_matrix=(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#changing estimator value to modify accuracy with same train n test split ratio
rfc1 = RandomForestClassifier(n_estimators = 50)
rfc1.fit(X_train, y_train)
```

```python
y_pred1=rfc1.predict(X_test)
confusion_matrix=(y_test,y_pred1)
accuracy_score(y_test,y_pred1)
```

```python
'''
    To improve model accuracy, considering only few main features and predicting the accuracy.
    We going to use only 0,1,2,3,5 columns as features to modify the accuracy of the model.
'''
X_categoric2 = df1.iloc[:, [1,2,3]].values
one = OneHotEncoder()
categoric_data = one.fit_transform(X_categoric2).toarray()
categoric_df2 = pd.DataFrame(categoric_data)
categoric_df2.columns = one.get_feature_names_out()
categoric_df2.columns
```

```python
X_final2 = pd.concat([numeric_df, categoric_df2], axis = 1)
```

```python
rfc2 = RandomForestClassifier(n_estimators = 100)
rfc2.fit(X_final2, y)
X_train2, X_test2, y_train2, y_test2=train_test_split(X_final2,y, test_size= 0.2)
rfc2.fit(X_train2, y_train2)
```

```python
y_pred2=rfc2.predict(X_test2)
confusion_matrix=(y_test2,y_pred2)
accuracy_score(y_test2,y_pred2)
```
